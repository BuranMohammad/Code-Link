```python
import csv
import networkx as nx
import numpy as np
from operator import itemgetter 
from scipy.stats import kendalltau
import pandas as pd
import math
from numpy import linalg as LA

G=nx.Graph()

f = open("bio-dmela.txt", "r")
edge1=[]
edge2=[]
nodelist=[]
for x in f:
  e1,e2=x.split()
  edge1.append(int(e1))
  edge2.append(int(e2))
  for i in (int(e1),int(e2)):
      if i not in nodelist:
          nodelist.append(i)
nodelist.sort()
t=0
nnodelist=[]
dic={}
for i in nodelist:
    dic[i]=t
    nnodelist.append(dic[i])
    t=t+1

nodelist=nnodelist
for  i in range(0, len(edge1)):
    G.add_edge(dic[edge1[i]], dic[edge2[i]])

nedge1=[]
nedge2=[]
for  i in range(0, len(edge1)):
   nedge1.append(dic[edge1[i]])
   nedge2.append(dic[edge2[i]])
   
edge1=nedge1
edge2=nedge2
        
n=len(nodelist)

d=[0]*n
for i in  range(0,n):
    d[i]=edge1.count(i)+edge2.count(i)
d=[val for (node,val) in G.degree()]
S_del=min(d)
B_del=max(d)
#print(d,S_del,B_del)


def LN(v,L):
    #Level=int(diameter()/2)
    neiL=[v]
    for i in range(0,L):
        for j in neiL:
            neiL=neiL+[n for n in G.neighbors(j)]
        neiL=list(set(neiL))
    return neiL

Dia2=int(nx.diameter(G)/2)
Dia=nx.diameter(G)
#para=3
def isc(hop):
    ISC_FN=[]
    for j in G.nodes():
            N=0
            for l in LN(j,int(hop)):
                 if G.degree(l)==S_del and l!=j:
                        N+=1
            ISC_FN.append(N*G.degree(j))
    return ISC_FN,G.nodes()

#print(isc(1))
#print(G.nodes())

   

    
# from operator import itemgetter
values,nodes=isc(1)
#print("befored sorted")
#print(values,nodes)
values,nodes=zip(*sorted(zip(values,nodes)))
values,nodes=list(values),list(nodes)
values.reverse()
nodes.reverse()
#print("after sorted")
print("Values: ",values)
print("Nodes :",nodes)
print("Top-10 Values: ",values[:10])
print("Top-10 Nodes: ",nodes[:10])
 


#nx.draw(G,with_labels=True)
#plt.show()
# for i in range(0,4):
#     print(LN(5,i))

    
n=10
x=[]
ISC_S_del=[]
#print(len(G.nodes))
for i in range(0,round(len(G.nodes())*0.75),n):
    new_G=nx.Graph()
    new_G = G.copy()
    new_G.remove_nodes_from(nodes[:i])
    #print(G.nodes)
    #print("Gnodes",len(G.nodes))
    #print("newGnodes",len(new_G.nodes))
    x.append(i)
    ISC_S_del.append(nx.number_connected_components(new_G))
    #print(i)
#print(x)
#print(ISC_D)
#print(len(nodes[:i]))

df=pd.DataFrame(list(zip(x,ISC_S_del)),columns=['No_of_Nodes_Removed','No_of_Connected_Components'])
print(df)
df.to_excel('ca_ISC_S_del_No_Connected_Components.xlsx','Sheet1')

df1=pd.DataFrame(list(zip(values,nodes)),columns=['Values','Nodes'])
print(df1)
df1.to_excel('ca_ISC_S_del_VN.xlsx','Sheet1')
```
