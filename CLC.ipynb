```python
import csv
import networkx as nx
import numpy as np
import pandas as pd
from operator import itemgetter 
from scipy.stats import kendalltau
G=nx.Graph()
import matplotlib.pyplot as plt
f = open("bio-dmela.txt", "r")
edge1=[]
edge2=[]
nodelist=[]
for x in f:
  e1,e2=x.split()
  edge1.append(int(e1)-1)
  edge2.append(int(e2)-1)
  for i in (int(e1)-1,int(e2)-1):
      if i not in nodelist:
          nodelist.append(i)
nodelist.sort()
t=0
nnodelist=[]
dic={}
for i in nodelist:
    dic[i]=t
    nnodelist.append(dic[i])
    t=t+1

nodelist=nnodelist
for  i in range(0, len(edge1)):
    G.add_edge(dic[edge1[i]], dic[edge2[i]])


CC=nx.clustering(G)
nodes = []
values = []

for item in CC.items():
    nodes.append(item[0]), values.append(item[1])
 
# printing keys and values separately

#print("befored sorted")
#print("values : ", values)
#print("Nodes : ", nodes)

values,nodes=zip(*sorted(zip(values,nodes)))
values,nodes=list(values),list(nodes)
values.reverse()
nodes.reverse()
#print("after sorted")
print("Values: ", values)
print("Nodes: ", nodes)
print("Top-10 Values: ",values[:10])
print("Top-10 Nodes: ",nodes[:10])
    
    
# n=10
# x=[]
# CLC=[]
# #print(len(G.nodes))
# for i in range(0,round(len(G.nodes())*0.75),n):
#     new_G=nx.Graph()
#     new_G = G.copy()
#     new_G.remove_nodes_from(nodes[:i])
#     #print(G.nodes)
#     print("Gnodes",len(G.nodes))
#     print("newGnodes",len(new_G.nodes))
#     x.append(i)
#     l=nx.number_connected_components(new_G)
#     print(l)
#     CLC.append(nx.number_connected_components(new_G))
#     #print(i)
# #print(x)
# #print(CLC)
# #print(len(nodes[:i]))



# df=pd.DataFrame(list(zip(x,CLC)),columns=['No_of_Nodes_Removed','No_of_Connected_Components'])
# print(df)
# df.to_excel('US_CLC_No_Connected_Components.xlsx','Sheet1')


# df1=pd.DataFrame(list(zip(values,nodes)),columns=['Values','Nodes'])
# print(df1)
# df1.to_excel('US_CLC_VN.xlsx','Sheet1')
```
